```python
import pandas as pd
import numpy as nd
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.neighbors import NearestNeighbors
from sklearn.metrics import accuracy_score
```

```python
df=pd.read_csv('project_I.csv')
df
```

```python

df=df.drop(columns={'shop','email','order_id','created_at','device','price','api_key','recommended','id'})
```

```python
df
```

```python
encoded_df=pd.get_dummies(df,columns=['product'],dtype=int)
encoded_df
```

```python
ab=encoded_df.groupby('uid').agg('sum').iloc[:,10:]
ab
```

```python
from sklearn.model_selection import train_test_split
from sklearn.neighbors import NearestNeighbors
X_train, X_test = train_test_split( ab, test_size=0.003, random_state=42)
knn=NearestNeighbors(n_neighbors=5)
knn.fit(X_train)
```

```python
df1=knn.kneighbors(X_test, return_distance=False)
df1
```

```python
df2=knn.kneighbors(X_train,return_distance=False)
df2
```

```python
#for test set
display_test_data=[]
for i in range(len(df1)):
    display(X_test.index[i],df.iloc[df1[i],:][['product']].values)
```

```python
#for test set
display_test_data=[]
for i in range(len(df1)):
    display_test_data.append(df.iloc[df1[i],:][['product']].values)
display_test_data
```

```python
flat_test_data=[]
for arr in display_test_data:
    flattened_arr = [item for sublist in arr for item in sublist] #array flatten
    rows_df = pd.DataFrame([flattened_arr], columns=[f'Recommendation_{i+1}' for i in range(len(flattened_arr))])
    flat_test_data.append(rows_df)


df3 = pd.concat(flat_test_data, ignore_index=True)
print(df3)
```

```python
def take_product_rows(uid,df3,df4):
    values_df3=df3.loc[uid] if uid in df3.index else None
    vlaues_df4=df4.loc[uid] if uid in df4.index else None
    return values_df3,vlaues_df4

def user_input():
    uid=input('Please Write Your UID')
    values_df3,vlaues_df4=take_product_rows(uid,df3,df4)
    if values_df3 is not None:
        print("You may Also Like These Prodcuts:\n",values_df3)
    elif vlaues_df4 is not None:
        print("You may Also Like These Prodcuts:\n",vlaues_df4)
    else:
        print("No products found for UID:", uid, "in database")
user_input()
```
